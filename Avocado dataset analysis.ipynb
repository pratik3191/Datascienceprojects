# Data Visualization

```python
#Lets visualize the average prices
plt.figure(figsize=(10,8))
plt.title('Average Price Distribution')
sns.distplot(df['AveragePrice'])
```

```python
plt.figure(figsize = (20,10))
df['AveragePrice'].plot.bar()
plt.ylabel('Average Prices')
plt.xlabel('No of Rows')
plt.xticks(rotation='vertical')
plt.title('Average price Distribution')
plt.show()
```

```python
plt.figure(figsize=(16,10))
plt.title('Avocado price in different Region')
plt.xticks(rotation='vertical')
sns.boxplot(x= 'region', y = 'AveragePrice', data=df)
```

Observation :- Average price for avocado is maximum is in BuffalorRochester and minimum price is in PhoenixTucson
